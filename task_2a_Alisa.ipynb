```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

N = 50
a = 1
b = -0.01
x0 = 1
gammas = [10**i for i in range(-3, 1)]

c = np.array([a**i * x0 for i in range(N+1)])
S = np.zeros((N+1, N))
for i in range(N+1):
    for j in range(i):
        S[i][j] = a**(i-j-1)*b

fig_u = go.Figure()
fig_x = go.Figure()

us = []
for gamma in gammas:
    
    u = -np.linalg.inv(np.transpose(S)@S + gamma*np.eye(N)) @ np.transpose(S) @ c
    x = c + S @ u


    fig_u.add_trace(go.Scatter(y=u, mode='lines+markers', name=f'gamma={gamma}'))
    fig_x.add_trace(go.Scatter(y=x, mode='lines+markers', name=f'gamma={gamma}'))

fig_u.update_layout(
    title="Optimal Control Signals u*",
    xaxis_title="Time step i",
    yaxis_title="u*",
)
fig_u.show()

fig_x.update_layout(
    title="Optimal Trajectories x_u*",
    xaxis_title="Time step i",
    yaxis_title="x",
)
fig_x.show()
```
