```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
def phi(x, n):
    return np.array([x**i for i in range(n)])
```

```python
train_df = pd.read_csv("trainingIa.dat", delim_whitespace=True, header=None)

# optional: name the columns
train_df.columns = ["x", "V"]
x = train_df["x"]
V = train_df["V"]

val_df = pd.read_csv("validationIa.dat", delim_whitespace=True, header=None)
val_df.columns = ["x", "V"]
m = len(val_df)

print(train_df.head())
```

```python
n = 20
# Design matrix
X = np.array([phi(xi, n) for xi in x])
model = LinearRegression(fit_intercept=False)  # since your design matrix is custom
model.fit(X, V)

theta = model.coef_
type(theta)
```

```python
N = 20
MSE = [0] * N
for n in range(1, N+1):
    X = np.array([phi(xi, n) for xi in x])
    model = LinearRegression(fit_intercept=False)  # since your design matrix is custom
    model.fit(X, V)
    theta = model.coef_

    phi_val = np.array([phi(xi, n) for xi in val_df["x"]])
    V_pred = theta.T @ phi_val.T
    summ = 0
    MSE[n-1] = sum((V_pred[i] - val_df["V"][i]) ** 2 for i in range(m)) / m
```

```python
fig = go.Figure(data=go.Scatter(x=[i for i in range(N)], y = MSE, mode='lines+markers'))
fig.show()
```
