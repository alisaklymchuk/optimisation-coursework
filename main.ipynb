```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
def phi(x, n):
    return np.array([x**i for i in range(n)])
```

```python
df = pd.read_csv("trainingIa.dat", delim_whitespace=True, header=None)

# optional: name the columns
df.columns = ["x", "V"]
x = df["x"]
V = df["V"]

print(df.head())
```

```python
# Design matrix
X = np.array([phi(xi, 20) for xi in x])
model = LinearRegression(fit_intercept=False)  # since your design matrix is custom
model.fit(X, V)

beta = model.coef_
```
